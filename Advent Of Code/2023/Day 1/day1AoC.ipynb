Step by step solution:<br />
Get the numbers from the day1input.txt file <br />
The input will be a list of unformatted strings. <br />

```python
input = open("day1input.txt", "r").readlines()
input = [elve.strip() for elve in input]
print(input)
```

Note: words like 'twone' should end up as 2 and 1, not only 2. That's why we have to set regex `overlapped=True`

```python
import regex as re
input = [re.findall(r'\d|one|two|three|four|five|six|seven|eight|nine', x, overlapped=True) for x in input]
print(input)
```

```python
def parse_int(string):
    ONES = {'one': '1',
            'two': '2',
            'three': '3',
            'four': '4',
            'five': '5',
            'six': '6',
            'seven': '7',
            'eight': '8',
            'nine': '9'
            }
    return ONES[string]
```

```python
output = []
for i in input:
  if not i[0].isdigit():
    i[0] = parse_int(i[0])
  if not i[-1].isdigit():
    i[-1] = parse_int(i[-1])
  output.append(i[0] + i[-1])
output = list(map(int, output))
print(output)
```

```python
print(sum(output))
```

```python
input = open("input.txt", "r").readlines()
index = input.index("\n")
rules, updates = input[:index], input[index + 1 :]
# strip the newline character
rules = [rule.strip() for rule in rules]
# turn to int
rules = [[int(x) for x in rule.split("|")] for rule in rules]
print(rules)

updates = [row.strip() for row in updates]
updates = [[int(x) for x in row.split(",")] for row in updates]
print(updates)
```

```python
# PART 1
def is_ordered(ruleset, update):
    for rule in ruleset:
        first_index = update.index(rule[0])
        second_index = update.index(rule[1])
        if first_index > second_index:
            return False
    return True


# if both number in a rule is contained in the update, add the rule to the related_rules
def get_valid_rules(update, rules):
    return {tuple(rule) for rule in rules if all(num in update for num in rule)}


correct_middle_page_total = 0
for update in updates:
    related_rules = get_valid_rules(update, rules)
    if is_ordered(related_rules, update):
        # get the middle page number
        correct_middle_page_total += update[(len(update) - 1) // 2]

print(correct_middle_page_total)
```

```python
# PART 2
undordered_list = [
    (update, get_valid_rules(update, rules))
    for update in updates
    if not is_ordered(get_valid_rules(update, rules), update)
]


def order_page(ruleset: set[tuple], page: list[int]) -> list:
    for rule in ruleset:
        first_index = page.index(rule[0])
        second_index = page.index(rule[1])
        if first_index > second_index:
            page[first_index], page[second_index] = (
                page[second_index],
                page[first_index],
            )
    if not is_ordered(ruleset, page):
        order_page(ruleset, page)
    return page


incorrect_middle_page_total = 0
# we now need to order the unordered list
for update, related_rules in undordered_list:
    # print(update, related_rules)
    update = order_page(related_rules, update)
    # print(update)
    incorrect_middle_page_total += update[(len(update) - 1) // 2]
print(incorrect_middle_page_total)
```

Challenges from https://adventofcode.com/2024/day/1 <br />
Step by step solution:<br />
Get the numbers from the day1input.txt file <br />


````python
input = open("day1input.txt", "r").readlines()
input = [elve.strip().split() for elve in input]
print(input)
````

We use map to turn every str into int, then use zip() with the \* operator to unzip the input into multiple lists, then zip it back with: list1 contains all the item at index 0 of every list in input, list2 all items at index 1, etc.

```
x = [1, 2, 3]
y = [4, 5, 6]
zipped = zip(x, y)
list(zipped) ==> [(1, 4), (2, 5), (3, 6)]

x2, y2 = zip(\*zipped)
x == list(x2) and y == list(y2) ==> True
```


````python
list1, list2 = zip(*input)
print(list1)
print(list2)
````

We then need to turn the tuples of strings into lists of integers, then sort them


````python
list1 = list(map(int, list1))
list1.sort()
list2 = list(map(int, list2))
list2.sort()
print(list1)
print(list2)
````

For the actual task of part 1: sum of the different between each item in the 2 list


````python
sum = 0
for i, j in zip(list1, list2):
    diff = abs(i - j)
    sum += diff

print(sum)
````

Part 2: calculate the similarity score


````python
score = 0
for i in list1:
    occurence = list2.count(i)
    score += i * occurence
print(score)
````
